```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("assets\\csv_files\\output_T.csv")
```

```python
df.head()
```

```python
df1=df.copy()
df1.replace("--",np.nan,inplace=True)

```

```python
df1.dropna(axis=0,inplace=True)
```

```python
df1.head()
```

```python

# Convert the specified portion of the DataFrame to a list
l = list(df1.iloc[:, 3:-4].values)

# Iterate over rows and columns in the list
for row_idx, row in enumerate(l):
    for col_idx, val in enumerate(row):
        if "*" in str(val):
            # Replace the value and update it in the list
            l[row_idx][col_idx] = str(val).replace("*", "")

# Update the DataFrame with the cleaned values
df1.iloc[:, 3:-4] = l

# Print the cleaned list for verification
print(l)
```

```python

s=df1.iloc[:,3:-4].mean(axis=0).round(2)
marks_average=s.reset_index()
marks_average.columns=["Course_Code","Average_Marks"]
```

```python
marks_average
```

```python

# Plotting the bar graph using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Course_Code', y='Average_Marks', data=marks_average, palette='viridis', edgecolor='black')

# Adding labels and title
plt.xlabel('Course Code', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.title('Average Internal Marks ', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the graph
plt.tight_layout()
plt.show()
```
